## Part 1: What is a "Pip Package"?
In Python, you don't have to write everything from scratch.

Most Windows users are used to "Plug and Play." When you plug in a USB stick, Windows automatically "mounts" it—assigning it a drive letter like `D:`.

In a cloud environment like Google Colab, the computer you are using is a temporary "virtual machine" in a Google data center. When you close the tab, that computer is wiped clean.
**"Mounting Google Drive"** is like virtually plugging your personal USB hard drive into that cloud computer so you can save your files permanently.

## Part 3: PyTorch Basics
PyTorch is the "engine" we will use for Deep Learning in Assignment 2 and 3.
While standard Python uses **Lists**, PyTorch uses **Tensors**.
Think of a **Tensor** as a grid of numbers (like a Sudoku board or an Excel sheet) that is specifically designed to be processed by a GPU (Graphics Card).

```python
import torch

# Create a 'Tensor' (a math-ready array)
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)

print(f"Our Tensor:\n{x_data}")
print(f"Shape of the grid: {x_data.shape}")

# Moving the math to the GPU (The 'Turbo' button for AI)
if torch.cuda.is_available():
    x_data_gpu = x_data.to("cuda")
    print("The tensor has been moved to the GPU for high-speed math!")
else:
    print("GPU not detected. Staying on the CPU.")
```
